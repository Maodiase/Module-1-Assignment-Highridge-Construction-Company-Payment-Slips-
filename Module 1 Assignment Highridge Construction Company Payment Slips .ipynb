```python
import random

#Name pools
MALE_FIRST_NAMES = [
    "James", "John", "Robert", "Michael", "William", "David", "Richard",
    "Joseph", "Thomas", "Charles", "Christopher", "Daniel", "Matthew",
    "Anthony", "Mark", "Donald", "Steven", "Paul", "Andrew", "Joshua",
    "Kevin", "Brian", "George", "Timothy", "Ronald", "Edward", "Jason",
    "Jeffrey", "Ryan", "Jacob", "Gary", "Nicholas", "Eric", "Jonathan",
    "Stephen", "Larry", "Justin", "Scott", "Brandon", "Frank",
]

FEMALE_FIRST_NAMES = [
    "Mary", "Patricia", "Jennifer", "Linda", "Barbara", "Elizabeth",
    "Susan", "Jessica", "Sarah", "Karen", "Lisa", "Nancy", "Betty",
    "Margaret", "Sandra", "Ashley", "Dorothy", "Kimberly", "Emily",
    "Donna", "Michelle", "Carol", "Amanda", "Melissa", "Deborah",
    "Stephanie", "Rebecca", "Sharon", "Laura", "Cynthia", "Kathleen",
    "Amy", "Angela", "Shirley", "Anna", "Brenda", "Pamela", "Emma",
    "Nicole", "Helen", "Diane",
]

LAST_NAMES = [
    "Smith", "Johnson", "Williams", "Brown", "Jones", "Garcia", "Miller",
    "Davis", "Rodriguez", "Martinez", "Hernandez", "Lopez", "Gonzalez",
    "Wilson", "Anderson", "Thomas", "Taylor", "Moore", "Jackson", "Martin",
    "Lee", "Perez", "Thompson", "White", "Harris", "Sanchez", "Clark",
    "Ramirez", "Lewis", "Robinson", "Walker", "Young", "Allen", "King",
    "Wright", "Scott", "Torres", "Nguyen", "Hill", "Flores", "Green",
    "Adams", "Nelson", "Baker", "Hall", "Rivera", "Campbell", "Mitchell",
    "Carter", "Roberts",
]


# Worker generation
def generate_workers(count: int = 400) -> list:
    """Dynamically build a list of worker records."""
    random.seed(42)          # fixed seed → reproducible results
    workers = []
    for i in range(1, count + 1):
        gender = random.choice(["Male", "Female"])
        first  = (
            random.choice(MALE_FIRST_NAMES)
            if gender == "Male"
            else random.choice(FEMALE_FIRST_NAMES)
        )
        last   = random.choice(LAST_NAMES)
        salary = round(random.uniform(5_000, 35_000), 2)
        workers.append({
            "id":     f"EMP{i:04d}",
            "name":   f"{first} {last}",
            "gender": gender,
            "salary": salary,
        })
    return workers

```

```python
# Employee level logic
def get_employee_level(salary: float, gender: str) -> str:
    """
    Determine employee level from salary + gender.
      A1   : $10,000 < salary < $20,000  (any gender)
      A5-F : $7,500  < salary < $30,000  AND Female
             (A5-F overrides A1 when both conditions met)
      Standard: no condition satisfied
    """
    level = "Standard"

    if 10_000 < salary < 20_000:          
        level = "A1"

    if 7_500 < salary < 30_000 and gender == "Female":  
        level = "A5-F"          

    return level
```

```python
def generate_payment_slip(worker: dict) -> str:
    """Format and return a payment slip for one worker."""
    level = get_employee_level(worker["salary"], worker["gender"])
    return (
        f"\n{'='*40}\n"
        f"{'PAYMENT SLIP':^40}\n"
        f"{'='*40}\n"
        f"  Employee ID    : {worker['id']}\n"
        f"  Name           : {worker['name']}\n"
        f"  Gender         : {worker['gender']}\n"
        f"  Salary         : ${worker['salary']:>10,.2f}\n"
        f"  Employee Level : {level}\n"
        f"{'='*40}\n"
    )
```

```python
def main():
    try:
        workers = generate_workers(400)

        if not workers:
            raise ValueError("Worker list is empty — no slips to generate.")

        print(f"Generating payment slips for {len(workers)} workers...\n")

        level_counts = {"A1": 0, "A5-F": 0, "Standard": 0}

        for worker in workers:
            try:
                slip  = generate_payment_slip(worker)
                print(slip)
                level = get_employee_level(worker["salary"], worker["gender"])
                level_counts[level] += 1

            except KeyError as e:
                print(f"[ERROR] Missing field {e} for worker. Skipping.")
            except Exception as e:
                print(f"[ERROR] Unexpected error for {worker.get('id','???')}: {e}. Skipping.")

        # ── Summary ──
        print(f"\n{'='*40}")
        print(f"{'SUMMARY':^40}")
        print(f"{'='*40}")
        print(f"  Total workers processed      : {len(workers)}")
        print(f"  Level A1  (10k–20k)         : {level_counts['A1']}")
        print(f"  Level A5-F (female 7.5k–30k): {level_counts['A5-F']}")
        print(f"  Level Standard               : {level_counts['Standard']}")
        print(f"{'='*40}\n")

    except ValueError as e:
        print(f"[CRITICAL] Data error: {e}")
    except Exception as e:
        print(f"[CRITICAL] Unexpected error: {e}")


if __name__ == "__main__":
    main()
```
